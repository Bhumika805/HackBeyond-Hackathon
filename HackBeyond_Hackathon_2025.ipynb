```python
yaml_text = """
path: /content/dataset
train: images/train
val: images/val

names:
  0: class0
  1: class1
  2: class2
  3: class3
  4: class4
  5: class5
  6: class6
"""

with open("config.yaml", "w") as f:
    f.write(yaml_text)
```

```python
!pip install ultralytics
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="config.yaml",
    epochs=30,
    imgsz=640,
    batch=8
)
```

```python
metrics = model.val()
metrics
```
